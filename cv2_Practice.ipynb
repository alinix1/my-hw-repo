First argument is to use the function cv2.imread() to read an image. The image should be in the working directory or a full path of image should be given.

Second argument is a flag which specifies the way image should be read.


    cv2.IMREAD_COLOR : Loads a color image. Any transparency of image will be neglected. It is the default flag.

    cv2.IMREAD_GRAYSCALE : Loads image in grayscale mode

    cv2.IMREAD_UNCHANGED : Loads image as such including alpha channel

Or use 1, 0, -1 to indicate the above arguments.

### Preprocessing a single image


```python
img_path = '/Users/alicevnix/Desktop/pic/frog0.jpg'
```

```python
from keras.preprocessing import image
import numpy as np
```

```python
img = image.load_img(img_path, target_size=(180, 180)) 
img_tensor = image.img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis=0)
img_tensor /= 255.
```

```python
print(img_tensor.shape)
```

## Display the Image of a Frog - using cv2

```python
import os.path
import numpy as np
import cv2

# construct an absolute path to the directory this file is located in 
HERE = os.path.dirname(os.path.abspath('/Users/alicevnix/Desktop/pic/frog1.jpg'))
```

```python
image_path = os.path.join(HERE, 'frog1.jpg')
img = cv2.imread(image_path, cv2.IMREAD_COLOR)
print(img)
```

```python
cv2.imshow('Test image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

cv2.waitKey( ) is a keyboard binding function. Its argument is the time in milliseconds. The function waits for specified milliseconds for any keyboard event. If you press any key in that time, the program continues. If 0 is passed, it waits indefinitely for a key stroke.

cv2.destroyAllWindows( ) simply destroys all the windows we created. If you want to destroy any specific window, use the function cv2.destroyWindow() where you pass the exact window name as the argument.
